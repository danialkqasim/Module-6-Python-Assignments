## **13.1**
> Write the current date as a string to the text file `today.txt`.
***

```python
from datetime import date
from pathlib import Path

# get date and write to file
now = date.today()
Path('today.txt').write_text(str(now))

print(f"Current date saved: {now}")
```

***
## **13.2**
> Read the text file `today.txt` into the string `today_string`.
***

```python
from pathlib import Path

# read file into variable
today_string = Path('today.txt').read_text()

print(f"today_string contents: {today_string}")
```

***
## **13.3**
> Parse the date from `today_string`.
***

```python
from datetime import datetime

# parse string into date object
fmt = '%Y-%m-%d'
parsed_date = datetime.strptime(today_string, fmt).date()

print(f"Parsed object: {parsed_date}")
```

## **15.1**
> Use `multiprocessing` to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit.
***

```python
import multiprocessing
import random
import time
from datetime import datetime

def task():
    # wait random time then print
    time.sleep(random.random())
    print(f"Process {multiprocessing.current_process().name} | Time: {datetime.now()}")

if __name__ == "__main__":
    # create and start 3 processes
    # output might show in terminal/console
    for _ in range(3):
        p = multiprocessing.Process(target=task)
        p.start()
```
